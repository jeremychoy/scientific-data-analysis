```python
import numpy as np
import csv

import matplotlib.pyplot as plt

# change the font globally
plt.rc("font", size=16, family='serif')

with open('../iris/iris_dataset.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)

data_array = np.array(data)

column_names = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm", "Species"]

species_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

masks = [data_array[:, 4] == name for name in species_names]

datasets = [np.asarray(data_array[mask][:,0:4], dtype=np.float32) for mask in masks]
```

```python
# Plot the datasets using the techniques of Chapter 5, Section 8

index_x = 0 # Sepal length
index_y = 2 # Petal length

x_data = [dataset[index_x] for dataset in datasets]
y_data = [dataset[index_y] for dataset in datasets]

fig, ax = plt.subplots(1, 3, sharey=True, figsize=(18,6))

for plot_number in range(3):
    ax[plot_number].plot(x_data[plot_number],
                         y_data[plot_number],
                         '.',
                         label=species_names[plot_number])
    ax[plot_number].set_xlabel(column_names[index_x])

ax[0].set_ylabel(column_names[index_y])

# plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.04, hspace=None)

plt.show()
```
